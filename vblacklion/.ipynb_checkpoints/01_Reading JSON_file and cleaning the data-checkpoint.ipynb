```python
def df_text_cleaner(df):
    """
    Takes a df as argument and cleans emptly lists as well as next_page strings
    
    'comfort&convenience', 'extras'
    """
    
    cln_dict = {}
    for name in df.columns:
       
        items = []
        for idx, item in df[[name]].iterrows():
            
            if type(item[0]) == list:
                if len(item[0]) == 1:
                    cln_item = item[0][0].replace('\n', '')
                    items.append(cln_item)
                elif len(item[0]) > 1:
                    list_item = []
                    for i in range(len(item[0])):
                        if type(item[0][i]) == list:
                            for y in range(len(item[0][0])):
                                cln_item = item[0][y].replace('\n', '')
                                list_item.append(cln_item)
                            if len(list_item) >= 1:
                                items.append(list_item)
                            else:
                                items.append(np.nan)
                        else:
                            for z in range(len(item[0])):
                                cln_item = item[0][z].replace('\n', '')
                                list_item.append(cln_item)
                            items.append(list_item)
                else:
                    cln_item = np.nan
                    items.append(cln_item)
            else:
                items.append(item[0])
        


        cln_dict[name] = items        
        
    return cln_dict

df_dict = df_text_cleaner(df_e)
```
